```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Directories
DATA_DIR = Path("../data")
FIG_DIR = Path("../figures")
FIG_DIR.mkdir(exist_ok=True)

# Load example data
df = pd.read_csv(DATA_DIR / "Table2_Gxs.csv")
x = df.iloc[:, 0]
y = df.iloc[:, 1]

# Create figure and axis
fig, ax = plt.subplots(figsize=(6, 4))

# Plot data
ax.plot(x, y, 's-', lw=2, markersize=5)

# -------------------------
# Axis control
# -------------------------

# Remove top/right/bottom spines except where needed
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(True)    # top X-axis for negative y
ax.spines['left'].set_visible(True)   # left Y-axis

# Place x-axis ticks on top
ax.xaxis.set_ticks_position('top')

# Place y-axis ticks on left only
ax.yaxis.set_ticks_position('left')

# If y is negative, reverse the y-axis so it grows downward
if np.any(y < 0):
    ax.invert_yaxis()   # makes negative y increase downward

# Tick direction inward
ax.tick_params(direction='in', length=6, width=1.2)

# Labels
ax.set_xlabel("Composition", labelpad=10)
ax.set_ylabel(r"$G^{xs}$ (J/mol)", labelpad=10)

# Grid off (or customize)
ax.grid(False)

# Tight layout
fig.tight_layout()

# Save figure
fig.savefig(FIG_DIR / "gxs_custom_axis.png", dpi=300, bbox_inches="tight")
plt.show()
```
