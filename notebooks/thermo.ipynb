```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Define paths (robust & clean)
DATA_DIR = Path("../data")
FIG_DIR = Path("../figures")

# Ensure figures directory exists
FIG_DIR.mkdir(exist_ok=True)

# Load CSV
df = pd.read_csv(DATA_DIR / "Table1_RK_parameters.csv")

print(df.head())
plt.figure(figsize=(6,4))

plt.plot(df.iloc[:,0], df.iloc[:,1], 'o-', label="RK data")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("R-K Parameters of Li-Mg Alloy")
plt.legend()
plt.grid(False)
plt.tight_layout()
# Save figure
plt.savefig(FIG_DIR / "rk_parameters.png", dpi=300)
plt.show()
```
