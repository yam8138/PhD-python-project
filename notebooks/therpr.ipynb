```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# APS-style defaults
plt.rcParams.update({
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 6,
    "ytick.major.size": 6,
    "axes.linewidth": 1.2,
    "font.size": 12,
})
DATA_DIR = Path("../data")
FIG_DIR  = Path("../figures")

FIG_DIR.mkdir(exist_ok=True)
# Load data
df = pd.read_csv(DATA_DIR / "Table1_RK_parameters.csv")

x = df.iloc[:, 0]
y = df.iloc[:, 1]

# Plot
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(x, y, 'o-', lw=2, markersize=5)

ax.set_xlabel("Composition")
ax.set_ylabel("RK parameter")

ax.tick_params(direction='in')
ax.minorticks_off()
ax.grid(False)

fig.tight_layout()
fig.savefig(FIG_DIR / "rk_parameters.png", dpi=300, bbox_inches="tight")
plt.show()
df = pd.read_csv(DATA_DIR / "Table2_Gxs.csv")

x = df.iloc[:, 0]
gxs = df.iloc[:, 1]

fig, ax = plt.subplots(figsize=(6,4))

ax.plot(x, gxs, 's-', lw=2, markersize=5)

ax.set_xlabel("Composition")
ax.set_ylabel(r"$G^{xs}$ (J/mol)")

ax.tick_params(direction='in', top=True, right=True)
ax.grid(False)

fig.tight_layout()
fig.savefig(FIG_DIR / "gxs_vs_composition.png", dpi=300, bbox_inches="tight")
plt.show()
df = pd.read_csv(DATA_DIR / "Table3_HM.csv")

x = df.iloc[:, 0]
hm = df.iloc[:, 1]

fig, ax = plt.subplots(figsize=(6,4))

ax.plot(x, hm, '^-', lw=2, markersize=5)

ax.set_xlabel("Composition")
ax.set_ylabel(r"$H_M$ (kJ/mol)")

ax.tick_params(direction='in')
ax.grid(False)

fig.tight_layout()
fig.savefig(FIG_DIR / "enthalpy_mixing.png", dpi=300, bbox_inches="tight")
plt.show()

```
