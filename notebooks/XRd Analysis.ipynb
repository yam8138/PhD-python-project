```python
# Example - this WILL work in your notebooks:
import numpy as np
import matplotlib.pyplot as plt

# XRD analysis
wavelength = 1.5406
two_theta = np.array([31.5, 45.2, 56.8])  # Your data
theta = np.radians(two_theta / 2)
d_spacings = wavelength / (2 * np.sin(theta))

print("d-spacings:", d_spacings)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(two_theta, d_spacings, 'o-')
plt.xlabel('2θ (degrees)')
plt.ylabel('d-spacing (Å)')
plt.title('XRD Analysis')
plt.grid(True)
plt.show()
```
