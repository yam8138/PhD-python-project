```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

plt.rcParams.update({
    "font.size": 12,
    "axes.linewidth": 1.2,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 6,
    "ytick.major.size": 6,
})
DATA_DIR = Path("../data")
FIG_DIR = Path("../figures")
FIG_DIR.mkdir(exist_ok=True)
df = pd.read_csv(DATA_DIR / "Table2_Gxs.csv")

x = df.iloc[:, 0].values
y = df.iloc[:, 1].values
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y, 'o-', lw=2, markersize=5)

# Labels
ax.set_xlabel("Composition")
ax.set_ylabel(r"$G^{xs}$ (J/mol)")

# Axis behavior (CRITICAL)
ax.set_ylim(y.min() * 1.1, y.max() * 1.1)   # ensures correct sign direction
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Ticks: ONLY left & bottom
ax.tick_params(
    axis='both',
    which='major',
    direction='in',
    top=False,
    right=False
)

ax.minorticks_off()
ax.grid(False)

fig.tight_layout()
fig.savefig(FIG_DIR / "gxs_vs_composition.png", dpi=300, bbox_inches="tight")
plt.show()
```
